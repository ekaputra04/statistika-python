```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset
file_path = 'nyc-east-river-bicycle-counts.csv'  # Ganti dengan path ke file CSV Anda
data = pd.read_csv(file_path)

# Membersihkan data
data['Date'] = pd.to_datetime(data['Date'])
data['Precipitation'] = data['Precipitation'].replace({'T': 0, '0.47 (S)': 0.47}).astype(float)
```

```python
# Mengelompokkan data berdasarkan bulan untuk analisis musiman
data['Month'] = data['Date'].dt.to_period('M')

# Menghitung rata-rata jumlah sepeda per bulan untuk setiap jembatan
bridges = ['Brooklyn Bridge', 'Manhattan Bridge', 'Williamsburg Bridge', 'Queensboro Bridge']
monthly_data = data.groupby('Month')[bridges].mean()

# Visualisasi tren musiman
plt.figure(figsize=(14, 7))
for bridge in bridges:
    plt.plot(monthly_data.index.to_timestamp(), monthly_data[bridge], label=bridge)

plt.xlabel('Month')
plt.ylabel('Average Number of Bicycles')
plt.title('Average Monthly Number of Bicycles per Bridge')
plt.legend()
plt.show()
```

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
file_path = 'nyc-east-river-bicycle-counts.csv'  # Ganti dengan path ke file CSV Anda
data = pd.read_csv(file_path)

# Membersihkan data
data['Date'] = pd.to_datetime(data['Date'])
data['Precipitation'] = data['Precipitation'].replace({'T': 0, '0.47 (S)': 0.47}).astype(float)

# Jembatan yang dianalisis
bridges = ['Brooklyn Bridge', 'Manhattan Bridge', 'Williamsburg Bridge', 'Queensboro Bridge']

# Analisis Tren Harian

# Visualisasi tren harian
plt.figure(figsize=(14, 7))
for bridge in bridges:
    plt.plot(data['Date'], data[bridge], marker='o', label=bridge)

plt.xlabel('Date')
plt.ylabel('Number of Bicycles')
plt.title('Daily Number of Bicycles Crossing Each Bridge')
plt.legend()
plt.xticks(rotation=45)  # Rotasi label tanggal jika perlu
plt.tight_layout()  # Mengoptimalkan tata letak plot
plt.show()
```
