# Soal 1

The following data relate x,the moisture of a wet mix of a certain product, to Y, the density of the finished product.

```
x = 5, 6, 7, 10, 12, 15, 18, 20
y = 7.4, 9.3, 10.6, 15.4, 18.1, 22.2, 24.1, 24.8
```

### a) Draw a scatter diagram

````python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

````python
x = np.array([5, 6, 7, 10, 12, 15, 18, 20]).reshape(-1, 1)
y = np.array([7.4, 9.3, 10.6, 15.4, 18.1, 22.2, 24.1, 24.8])

# Plot data asli dan garis regresi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Scatter Plot")
plt.legend()
plt.show()
````

### b) Fit a linear curve to the data.

````python
# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(x, y)

# Melakukan prediksi
y_pred = model.predict(x)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien: {model.coef_}")
print(f"Intersep: {model.intercept_}")

# Plot data asli dan garis regresi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Garis Regresi')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Regresi Linear")
plt.legend()
plt.show()
````
