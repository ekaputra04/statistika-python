````python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

Inisialisasi variabel

````python
locations_all = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Dist of Columbia", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico", "NewYork", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming", "National Average"]

math_all = np.array([558, 513, 521, 550, 511, 538, 504, 495, 473, 496, 477, 510, 536, 575, 494, 600, 571, 544, 550, 498, 504, 504, 565, 593, 557, 569, 547, 568, 507, 514, 505, 548, 499, 486, 599, 535, 557, 521, 492, 491, 474, 566, 552, 500, 575, 500, 496, 519, 506, 586, 544, 508]).reshape(-1, 1)

percentage_all = np.array([8, 47, 28, 6, 45, 30, 79, 66, 50, 48, 63, 54, 15, 14, 57, 5, 9, 12, 9, 68, 64, 80, 11, 9, 4, 9, 21, 9, 31, 70, 69, 12, 73, 59, 5, 24, 8, 50, 71, 69, 57, 5, 14, 48, 4, 70, 68, 47, 17, 8, 11, 41])
````

#### First 20 data

````python
first_20_location = locations_all[:20]
first_20_math = math_all[:20]
first_20_percentage = percentage_all[:20]

mean_20_math = np.mean(first_20_math)
print(f"Mean for 20 first data = {mean_20_math}")
````

Data visualitation for first 20 data

````python
# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(first_20_math, first_20_percentage)

# Melakukan prediksi
y_pred = model.predict(first_20_math)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien: {model.coef_}")
print(f"Intersep: {model.intercept_}")

plt.figure(figsize=(10, 6))
plt.scatter(first_20_math, first_20_percentage, color='blue', label='Actual data')
plt.plot(first_20_math, model.predict(first_20_math), color='red', label='Regression line')
plt.xlabel('Math Score')
plt.ylabel('Percentage')
plt.title("Regresi Linear")
plt.legend()
plt.grid()
plt.show()
````
