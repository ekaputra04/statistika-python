# Soal 2

The following data relate the number of units of a good that were ordered as a function of the price of the good at six different locations.

```
Number ordered = 88, 112, 123, 136, 158, 172
Price = 50, 40, 35, 30, 20, 15
```

How many units do you think would be ordered if the price were 25?

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
prices = np.array([50, 40, 35, 30, 20, 15]).reshape(-1, 1)
units_ordered = np.array([88, 112, 123, 136, 158, 172])

# Create a linear regression model
model = LinearRegression()
model.fit(prices, units_ordered)

# Predict the number of units for the price of 25
price_to_estimate = np.array([[25]])
units_estimated = model.predict(price_to_estimate)[0]

# Print the estimated number of units
print(f"Estimated units ordered at price {price_to_estimate[0][0]}: {units_estimated}")

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(prices, units_ordered, color='blue', label='Actual data')
plt.plot(prices, model.predict(prices), color='red', label='Regression line')
plt.scatter(price_to_estimate, units_estimated, color='green', label=f'Estimated at price {price_to_estimate[0][0]}', zorder=5)
plt.title('Number of Units Ordered vs. Price')
plt.xlabel('Price')
plt.ylabel('Number of Units Ordered')
plt.legend()
plt.grid(True)
plt.show()
````
