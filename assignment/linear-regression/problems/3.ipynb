# Soal 3

The corrosion of a certain metallic substance has been studied in dry oxygen at 500 degrees centigrade.In this experiment, the gain in weight after various periods of exposure was used as a measure of the amount of oxygen that had reacted with the sample. Here are the data:

```
Hours = 1.0, 2.0, 2.5, 3.0, 3.5, 4.0
Percent Gain = 0.02, 0.03, 0.035, 0.042, 0.05, 0.054
```

### a) Plot a scatter diagram

````python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

````python
x = np.array([1.0, 2.0, 2.5, 3.0, 3.5, 4.0]).reshape(-1, 1)
y = np.array([0.02, 0.03, 0.035, 0.042, 0.05, 0.054])

plt.scatter(x, y, color='blue', label='Data Asli')
plt.xlabel('Hours')
plt.ylabel('Percent Gain')
plt.title("Scatter Plot")
plt.legend()
plt.show()
````

### b) Fit a linear relation

````python
# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(x, y)

# Melakukan prediksi
y_pred = model.predict(x)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien: {model.coef_}")
print(f"Intersep: {model.intercept_}")

# Plot data asli dan garis regresi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Garis Regresi')
plt.xlabel('Hours')
plt.ylabel('Percent Gain')
plt.title("Regresi Linear")
plt.legend()
plt.show()
````

### c) Predict the percent weight gain when the metal is exposed for 3.2 hours.

````python
# Predict the number of units for the price of 25
hours_to_estimate = np.array([[3.2]])
units_estimated = model.predict(hours_to_estimate)[0]

print(f"Estimated percent gain at {hours_to_estimate[0][0]} hours : {units_estimated}")

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, model.predict(x), color='red', label='Regression line')
plt.scatter(hours_to_estimate, units_estimated, color='green', label=f'Estimated at {hours_to_estimate[0][0]} hours', zorder=5)
plt.xlabel('Hours')
plt.ylabel('Percent Gain')
plt.title("Regresi Linear")
plt.legend()
plt.grid(True)
plt.show()
````
