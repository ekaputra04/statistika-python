The following data indicate the gain in reading speed versus the number of weeks in the program of 10 students in a speed-reading program.

```
Number of week = 2, 3, 8, 11, 4, 5, 9, 7, 5, 7
Speed gain (wods / min) = 21, 42, 102, 130, 52, 57, 105, 85, 62, 90
```

### a) Plot a scatter diagram to see if a linear relationship is indicated.

````python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

````python
x = np.array([2, 3, 8, 11, 4, 5, 9, 7, 5, 7]).reshape(-1, 1)
y = np.array([21, 42, 102, 130, 52, 57, 105, 85, 62, 90])

plt.scatter(x, y, color='blue', label='Data Asli')
plt.xlabel('Number of week')
plt.ylabel('Speed gain')
plt.title("Scatter Plot")
plt.legend()
plt.show()
````

### b) Find the least squares estimates of the regression coefficients.

````python
# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(x, y)

# Melakukan prediksi
y_pred = model.predict(x)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien: {model.coef_}")
print(f"Intersep: {model.intercept_}")
````

### c) Estimate the expected gain of a student who plans to take the program for 7 weeks.

````python
# Predict the number of units for the price of 25
week_to_estimate = np.array([[7]])
units_estimated = model.predict(week_to_estimate)[0]

print(f"Estimated speed gain at {week_to_estimate[0][0]} weeks : {units_estimated}")

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, model.predict(x), color='red', label='Regression line')
plt.scatter(week_to_estimate, units_estimated, color='green', label=f'Estimated at {week_to_estimate[0][0]} = {units_estimated:.2f}', zorder=5)
plt.xlabel('Number of weeks')
plt.ylabel('Speed gain')
plt.title("Regresi Linear")
plt.legend()
plt.grid()
plt.show()
````
