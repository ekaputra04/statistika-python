# Soal 4

The following data indicate the relationship between x, the specific gravity of a wood sample, and Y, its maximum crushing strength in compression parallel to the grain.

```
xi = 0.41, 0.46, 0.44, 0.47, 0.42, 0.39, 0.41, 0.44, 0.43, 0.44
yi = 1850, 2620, 2340, 2690, 2160, 1760, 2500, 2750, 2730, 3120
```

### a) Plot a scatter diagram. Does a linear relationship seem reasonable?

````python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

````python
x = np.array([0.41, 0.46, 0.44, 0.47, 0.42, 0.39, 0.41, 0.44, 0.43, 0.44]).reshape(-1, 1)
y = np.array([1850, 2620, 2340, 2690, 2160, 1760, 2500, 2750, 2730, 3120])

plt.scatter(x, y, color='blue', label='Data Asli')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title("Scatter Plot")
plt.legend()
plt.show()
````

### b) Estimate the regression coefficients.

````python
# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(x, y)

# Melakukan prediksi
y_pred = model.predict(x)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien: {model.coef_}")
print(f"Intersep: {model.intercept_}")

# Plot data asli dan garis regresi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Garis Regresi')
plt.xlabel('xi')
plt.ylabel('yi')
plt.title("Regresi Linear")
plt.legend()
plt.show()
````

### c) Predict the maximum crushing strength of a wood sample whose specific gravity is 0.43.

````python
# Predict the number of units for the price of 25
gravity_to_estimate = np.array([[0.43]])
units_estimated = model.predict(gravity_to_estimate)[0]

print(f"Estimated crushing strength at {gravity_to_estimate[0][0]} : {units_estimated}")

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, model.predict(x), color='red', label='Regression line')
plt.scatter(gravity_to_estimate, units_estimated, color='green', label=f'Estimated at {gravity_to_estimate[0][0]}', zorder=5)
plt.xlabel('Hours')
plt.ylabel('Percent Gain')
plt.title("Regresi Linear")
plt.legend()
plt.grid()
plt.show()
````
