Infrared spectroscopy is often used to determine the natural rubber content of mixtures of natural and synthetic rubber. For mixtures of known percentages, the infrared spectroscopy gave the following readings:

```
Percentage = 0, 20, 40, 60, 80, 100
Reading = 0.734, 0.885, 1.050, 1.191, 1.314, 1.432
```

If a new mixture gives an infrared spectroscopy reading of 1.15, estimate its percentage of natural rubber.

````python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
````

````python
x = np.array([0.734, 0.885, 1.050, 1.191, 1.314, 1.432]).reshape(-1, 1)
y = np.array([0, 20, 40, 60, 80, 100])

# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(x, y)

# Melakukan prediksi
y_pred = model.predict(x)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien: {model.coef_}")
print(f"Intersep: {model.intercept_}")

# Predict the number of units for the price of 25
reading_to_estimate = np.array([[1.15]])
units_estimated = model.predict(reading_to_estimate)[0]

print(f"Estimated speed gain at {reading_to_estimate[0][0]} weeks : {units_estimated}")

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, model.predict(x), color='red', label='Regression line')
plt.scatter(reading_to_estimate, units_estimated, color='green', label=f'Estimated at {reading_to_estimate[0][0]} = {units_estimated:.2f}', zorder=5)
plt.xlabel('Reading')
plt.ylabel('Percentage')
plt.title("Regresi Linear")
plt.legend()
plt.grid()
plt.show()
````
