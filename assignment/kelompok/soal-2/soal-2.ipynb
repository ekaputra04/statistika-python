# Soal 2

Rata-rata nilai kuliah statistik diketahui 65 dengan standard deviasi 15.

a)	Jika diinginkan 15% murid mendapatkan nilai A dan diketahui distribusi nilai normal, berapakah batas bawah nilai agar mendapatkan A?

b)	Selanjutnya diinginkan yang mendapatkan nilai B adalah sebanyak 25%, berapakah batas bawah nilai B?

c)	Seandainya diinginkan yang tidak lulus paling banyak 25%, berapakah batas bawah agar siswa lulus?

### Jawaban

Import Library

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

Mendata semua variabel yang diketahui pada soal

```python
# Diberikan
mean = 65      # Rata-rata nilai kuliah statistik
std_dev = 15   # Standar deviasi nilai kuliah statistik
```

## a) Batas bawah nilai agar mendapatkan A (15% murid)

```python
print("a) Batas bawah nilai agar mendapatkan A (15% murid):")
z_score_a = norm.ppf(0.85)  # Mencari z-score dengan proporsi kumulatif 0.85 (1 - 0.15)
lower_bound_a = mean + z_score_a * std_dev
print(f"Z score A: {z_score_a:.2f}")
print(f"Batas bawah nilai A: {lower_bound_a:.2f}")
```

Visualisasi distribusi normal dengan nilai z score > 15%

```python
# Membuat array x untuk visualisasi distribusi normal
x = np.linspace(-4, 4, 1000)
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Menghitung area di bawah kurva sebelah kiri dari z-score 1.04
area_left = np.trapz(y[:np.where(x <= z_score_a)[0][-1]+1], x[:np.where(x <= z_score_a)[0][-1]+1])
area_right = 1 - area_left

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'b-', lw=2, label='Distribusi Normal Standar')
ax.fill_between(x, 0, y, where=(x >= z_score_a), color='g', alpha=0.3, label=f'Area 15% (z = {z_score_a:.2f})')
ax.axvline(z_score_a, color='r', linestyle='--', label=f'z = {z_score_a:.2f}')
ax.legend(loc='upper left')
ax.set_title('Distribusi Normal Standar')
ax.set_xlabel('Nilai z')
ax.set_ylabel('Probabilitas Densitas')
ax.text(1.5, 0.35, f'Area kanan = {area_right:.2f} (15%)', fontsize=12)
ax.tick_params(axis='y', labelleft=False)  # Menyembunyikan angka pada sumbu y
plt.savefig("z-score-a.jpg")
plt.show()
```

Visualisasi distribusi normal dengan nilai murid > 15% adalah A

```python
# Memvisualisasikan distribusi normal
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
y = norm.pdf(x, loc=mean, scale=std_dev)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'b-', lw=2)
ax.fill_between(x, 0, y, where=(x >= lower_bound_a), color='g', alpha=0.3, label=f'15% siswa mendapat nilai A')
ax.axvline(lower_bound_a, color='r', linestyle='--', label=f'Batas bawah A ({lower_bound_a:.2f})')
ax.legend(loc='upper left')
ax.tick_params(axis='y', labelleft=False)  # Menyembunyikan angka pada sumbu y
ax.set_title('Distribusi Normal Nilai Kuliah Statistik')
ax.set_xlabel('Nilai')
ax.set_ylabel('Probabilitas Densitas')
plt.savefig("nilai-a.jpg")
plt.show()
```

## b) Batas bawah nilai agar mendapatkan B (25% murid)

```python
print("\nb) Batas bawah nilai agar mendapatkan B (25% murid):")
z_score_b = norm.ppf(0.6)  # Mencari z-score dengan proporsi kumulatif 0.6 (1 - (0.15 + 0.25))
lower_bound_b = mean + z_score_b * std_dev
print(f"Z score B: {z_score_b:.2f}")
print(f"Batas bawah nilai B: {lower_bound_b:.2f}")
```

Visualisasi distribusi normal dengan nilai z score > 40% (didapat dari 15% + 25%)

```python
# Membuat array x untuk visualisasi distribusi normal
x = np.linspace(-4, 4, 1000)
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Menghitung area di bawah kurva sebelah kiri dari z-score 1.04
area_left = np.trapz(y[:np.where(x <= z_score_b)[0][-1]+1], x[:np.where(x <= z_score_b)[0][-1]+1])
area_right = 1 - area_left

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'b-', lw=2, label='Distribusi Normal Standar')
ax.fill_between(x, 0, y, where=(x >= z_score_b), color='g', alpha=0.3, label=f'Area 40%')
ax.axvline(z_score_b, color='r', linestyle='--', label=f'z = {z_score_b:.2f}')
ax.legend(loc='upper left')
ax.set_title('Distribusi Normal Standar')
ax.set_xlabel('Nilai z')
ax.set_ylabel('Probabilitas Densitas')
ax.text(1.5, 0.35, f'Area kanan = {area_right:.2f}', fontsize=12)
ax.tick_params(axis='y', labelleft=False)  # Menyembunyikan angka pada sumbu y
plt.savefig("z-score-b.jpg")
plt.show()
```

Visualisasi distribusi normal dengan nilai murid > 25% adalah B

```python
# Memvisualisasikan distribusi normal
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
y = norm.pdf(x, loc=mean, scale=std_dev)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'b-', lw=2)
ax.fill_between(x, 0, y, where=((x >= lower_bound_b) & (x <= lower_bound_a)), color='g', alpha=0.3, label=f'25% siswa mendapat nilai B')
ax.axvline(lower_bound_b, color='r', linestyle='--', label=f'Batas bawah B ({lower_bound_b:.2f})')
ax.legend(loc='upper left')
ax.tick_params(axis='y', labelleft=False)  # Menyembunyikan angka pada sumbu y
ax.set_title('Distribusi Normal Nilai Kuliah Statistik')
ax.set_xlabel('Nilai')
ax.set_ylabel('Probabilitas Densitas')
plt.savefig("nilai-b.jpg")
plt.show()
```

## c) Batas bawah agar siswa lulus (maksimal 25% tidak lulus)
