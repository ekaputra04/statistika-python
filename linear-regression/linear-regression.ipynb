```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Contoh data: x (fitur/input) dan y (target/output)
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25])

# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(x, y)

# Melakukan prediksi
y_pred = model.predict(x)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien: {model.coef_}")
print(f"Intersep: {model.intercept_}")

# Plot data asli dan garis regresi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Garis Regresi')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Contoh data: x (fitur/input) dan y (target/output)
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25])

# Membuat model regresi linear
model = LinearRegression()

# Melatih model menggunakan data
model.fit(x, y)

# Membuat nilai x yang diperluas untuk memanjang garis regresi
x_extended = np.linspace(0, 6, 100).reshape(-1, 1)  # Dari 0 sampai 6

# Melakukan prediksi pada nilai x yang diperluas
y_extended_pred = model.predict(x_extended)

# Menampilkan koefisien dan intersep dari garis regresi
print(f"Koefisien (m): {model.coef_[0]}")
print(f"Intersep (c): {model.intercept_}")

# Plot data asli dan garis regresi yang diperluas
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x_extended, y_extended_pred, color='red', label='Garis Regresi Diperluas')

# Menyesuaikan batas sumbu x agar dimulai dari 0
plt.xlim(left=0)

# Menambahkan label dan legenda
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Menampilkan plot
plt.show()
```
